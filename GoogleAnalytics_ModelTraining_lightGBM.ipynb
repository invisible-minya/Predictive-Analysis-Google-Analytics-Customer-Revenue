```
# for return players label training 
lgb_model1 = lgb.LGBMClassifier(objective = 'binary',metric = 'binary_logloss')

params_opt = {'max_bin':[256], 'learning_rate':[0.01,0.02], 'num_leaves':[5,10,15], 'bagging_fraction':[0.8,0.9],'feature_fraction':[0.8,0.9],'min_data':[1,2], 'bagging_freq':[1,2]}
gridSearchCV = GridSearchCV(estimator = lgb_model1, param_grid = params_opt, cv = 3)
gridSearchCV.fit(X_train,y_train.loc[:,"will_return"])
gridSearchCV.best_estimator_, gridSearchCV.best_score_, gridSearchCV.best_params_
```
